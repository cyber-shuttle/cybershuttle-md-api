```python
from pathlib import Path

experiment_data_dir = Path("./outputs/ContinueReplica_namd_Run_on_Jun_8_2023_431/")
```

```python
from cybershuttle_md_cli import analysis
replica1_log = experiment_data_dir / "NAMD_Repl_1.out"
replica1_energies = analysis.load_energies(replica1_log)
replica2_log = experiment_data_dir / "NAMD_Repl_2.out"
replica2_energies = analysis.load_energies(replica2_log)
replica3_log = experiment_data_dir / "NAMD_Repl_3.out"
replica3_energies = analysis.load_energies(replica3_log)
```

```python
from matplotlib import pyplot as plt

fig,ax = plt.subplots(ncols=3,figsize=(15,5),
                      sharex=True,sharey=True)

ax[0].plot(replica1_energies['TS'],replica1_energies['Potential'])
ax[1].plot(replica2_energies['TS'],replica2_energies['Potential'])
ax[2].plot(replica3_energies['TS'],replica3_energies['Potential'])
ax[0].set_ylabel('Potential energy')

ax[0].set_xlabel('Timestep (replica 1)')
ax[1].set_xlabel('Timestep (replica 2)')
ax[2].set_xlabel('Timestep (replica 3)')
```
